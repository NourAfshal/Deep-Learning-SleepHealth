```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

label_encoder = LabelEncoder()
df['Gender_encoded'] =label_encoder.fit_transform(df['Gender'])
df['Occupation_encoded'] =label_encoder.fit_transform(df['Occupation'])
df['BMI_Category_encoded'] =label_encoder.fit_transform(df['BMI Category'])
df['Blood_Pressure_encoded'] =label_encoder.fit_transform(df['Blood Pressure'])
df['Sleep_Disorder_encoded'] =label_encoder.fit_transform(df['Sleep Disorder'])
```

```python
df_selected = df[['Gender_encoded', 'Occupation_encoded', 'BMI_Category_encoded',
                  'Blood_Pressure_encoded', 'Sleep_Disorder_encoded']]

df_selected.corr()
```

```python
person_id_column = df[['Person ID']]
gender_encoded_column = df_selected[['Gender_encoded']]
age_column = df[['Age']]
occupation_encoded_column = df_selected[['Occupation_encoded']]
sleep_columns = df[['Sleep Duration', 'Quality of Sleep', 'Physical Activity Level', 'Stress Level']]
bmi_blood_pressure_columns = df_selected[['BMI_Category_encoded', 'Blood_Pressure_encoded']]
heart_rate_daily_steps_columns = df[['Heart Rate', 'Daily Steps']]
sleep_disorder_column = df_selected['Sleep_Disorder_encoded']

result_df = pd.concat([person_id_column, gender_encoded_column, age_column,
                      occupation_encoded_column, sleep_columns, bmi_blood_pressure_columns,
                      heart_rate_daily_steps_columns, sleep_disorder_column], axis=1)

result_df
```

```python
result_df.corr()
```

```python
X_encoded = result_df.iloc[:, :-1].values  #Features all the columns except the last one
Y_encoded = result_df.iloc[:, -1].values   #label only last column
```

```python
X_encoded
```

```python
Y_encoded
```

```python
X_train_encoded, X_test_encoded, Y_train_encoded, y_test_encoded = train_test_split(X_encoded, Y_encoded, test_size=0.2, random_state=42)
```

```python
X_train_encoded
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_encoded = scaler.fit_transform(X_train_encoded)
X_test_encoded = scaler.transform(X_test_encoded)
```

```python
X_train_encoded
```

```python
X_test_encoded
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from  keras.models import Sequential
from  keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(32,  activation='relu', input_shape=(X_train_encoded.shape[1],)))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy', 'MSE', 'MAE'])
```

```python
from keras.utils import to_categorical
import numpy as np

model.fit(X_train_encoded, Y_train_encoded, epochs=50, batch_size=100, verbose=1)

```

```python
model.summary()
```

```python
from keras.utils import plot_model

plot_model(model, to_file='model_architecture.png', show_shapes=True)
```

```python
y_pred_prob = (model.predict(X_test_encoded)>0.5).astype("int32")
y_pred = np.argmax(y_pred_prob, axis=1)
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test_encoded, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(confusion_matrix(y_test_encoded, y_pred), annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

```python
from sklearn.metrics import classification_report

print("\nClassification Report:")
print(classification_report(y_test_encoded, y_pred))
```
